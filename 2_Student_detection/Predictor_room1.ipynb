```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from operator import itemgetter
import ipywidgets as widgets
from ultralytics import YOLOv10
```

```python
model = YOLOv10("/content/drive/MyDrive/Final Model/Room1/best.pt")
```

```python
def predict_label_with_boxes(confidence,iou):
    img = plt.imread("/content/drive/MyDrive/Siam/TrainDatabase/1302 (images,labels)/images/20240821132030.jpg")
    results = model(source=img,
                    conf=confidence/100,
                    iou=iou/100)

     # Set the figure size before plotting
    fig, ax = plt.subplots(1, figsize=(12, 8))  # Adjust the figsize as needed
    ax.imshow(img)

    for result in results:
        if result.boxes:
            boxes = result.boxes.xyxy.tolist()  # xyxy format for drawing boxes
            clases = [int(cls) for cls in result.boxes.cls.tolist()]

            for box, cls in zip(boxes, clases):
                x_min, y_min, x_max, y_max = box
                rect = patches.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min,
                                         linewidth=1, edgecolor='r', facecolor='none')
                ax.add_patch(rect)
                plt.text(x_min, y_min, f'', color='white', fontsize=12,
                         bbox=dict(facecolor='red', alpha=0.5))

    plt.show()

# Interactive widget for confidence slider
i_slider = widgets.IntSlider(value=15, min=1, max=100, step=3, description='confidence:')
iou_slider = widgets.IntSlider(value=20, min=0, max=100, step=5, description='IOU:')
widgets.interactive(predict_label_with_boxes, confidence=i_slider,iou=iou_slider)
```

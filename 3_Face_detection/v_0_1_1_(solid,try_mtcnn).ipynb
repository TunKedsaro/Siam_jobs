```python
from mtcnn.mtcnn import MTCNN
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/content/drive/MyDrive/Siam/Video_2_image_and_database/Source_video_from_siam/00000001572000000/ 122041.png")
image_rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
detector = MTCNN()
```

```python
faces = detector.detect_faces(image_rgb)
```

```python
faces
```

```python
for face in faces:
    print(face['box'])
    print(face['keypoints'])
```

```python
for face in faces:
    # Get the bounding box
    x, y, width, height = face['box']
    cv2.rectangle(image_rgb, (x, y), (x+width, y+height), (0, 255, 0), 2)  # Green bounding box

    # # Get keypoints (eyes, nose, mouth)
    # keypoints = face['keypoints']
    # cv2.circle(image_rgb, keypoints['left_eye'], 2, (0, 255, 255), 2)  # Yellow left eye
    # cv2.circle(image_rgb, keypoints['right_eye'], 2, (0, 255, 255), 2)  # Yellow right eye
    # cv2.circle(image_rgb, keypoints['nose'], 2, (255, 0, 0), 2)  # Blue nose
    # cv2.circle(image_rgb, keypoints['mouth_left'], 2, (0, 0, 255), 2)  # Red mouth (left)
    # cv2.circle(image_rgb, keypoints['mouth_right'], 2, (0, 0, 255), 2)  # Red mouth (right)

plt.imshow(image_rgb)
plt.axis('off')  # Hide axes
plt.show()
```
